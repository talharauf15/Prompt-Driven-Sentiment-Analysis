```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("TinyLlama/TinyLlama-1.1B-step-50K-105b")
model = AutoModelForCausalLM.from_pretrained("TinyLlama/TinyLlama-1.1B-step-50K-105b")
```

```python
def perform_sentiment_analysis(prompt):
    sentiment_prompt = f"The sentiment of this text is either positive, negative or neutral: {prompt} Sentiment:"

    inputs = tokenizer(sentiment_prompt, return_tensors="pt")

    outputs = model(**inputs)

    logits = outputs.logits

    # Extract the predicted sentiment class (positive, neutral, negative)
    # We will simply assume 3 possible outputs for simplicity
    predicted_class = torch.argmax(logits[:, -1, :], dim=-1).item()

    sentiment_map = {0: "negative", 1: "neutral", 2: "positive"}
    sentiment = sentiment_map.get(predicted_class, "neutral")  # Default to neutral if unknown

    return sentiment
```

```python
example = "I absolutely love this product. It's fantastic!"
sentiment = perform_sentiment_analysis(example)
print(f"Predicted Sentiment: {sentiment}")
```

```python
example = "This product is okay!"
sentiment = perform_sentiment_analysis(example)
print(f"Predicted Sentiment: {sentiment}")
```

```python
example = "This product is trash!"
sentiment = perform_sentiment_analysis(example)
print(f"Predicted Sentiment: {sentiment}")
```
