```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("openai-community/gpt2")
model = AutoModelForCausalLM.from_pretrained("openai-community/gpt2")
```

```python
import torch
```

```python
def perform_sentiment_analysis(prompt):
    sentiment_prompt = f"The sentiment of this text is either positive, negative or neutral: {prompt} Sentiment:"

    inputs = tokenizer(sentiment_prompt, return_tensors="pt") # return_tensors="pt",converts the tokenized input into a format that the PyTorch model can accept and process

    outputs = model.generate(
        inputs.input_ids,
        max_length=50,
        temperature=0.0,  # It gives deterministic, predictable output
        pad_token_id=tokenizer.eos_token_id
    )

    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    if "positive" in generated_text.lower():
        sentiment = "positive"
    elif "negative" in generated_text.lower():
        sentiment = "negative"
    else:
        sentiment = "neutral"

    return sentiment
```

```python
example = "I absolutely love this product. It's fantastic!"
sentiment = perform_sentiment_analysis(example)
print(f"Predicted Sentiment: {sentiment}")
```

```python
example = "this product is okey!"
sentiment = perform_sentiment_analysis(example)
print(f"Predicted Sentiment: {sentiment}")
```

```python
example = "This product is trash!"
sentiment = perform_sentiment_analysis(example)
print(f"Predicted Sentiment: {sentiment}")
```
